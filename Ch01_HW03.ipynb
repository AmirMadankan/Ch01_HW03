```python
import numpy as np
import matplotlib.pyplot as plt
from openpyxl import load_workbook
from sklearn.linear_model import LinearRegression
```

```python
ws = load_workbook("Ch01_HW03_linear_regression_data.xlsx")["Sheet1"]
```

```python
n = ws.max_row + 1
x = []
y = []
for i in range(2, n):
    x.append([ws[f"A{i}"].value, ws[f"B{i}"].value, ws[f"C{i}"].value])
    y.append([ws[f"D{i}"].value])
```

```python
x = np.array(x).reshape(-1, 3)
y = np.array(y).reshape(-1, 1)
```

```python
print(x)
```

```python
print(y)
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
plt.plot(x[:, 0], y, 'o')
plt.xlabel("input 1")
plt.ylabel("output")
```

```python
plt.plot(x[:, 1], y, 'o')
plt.xlabel("input 2")
plt.ylabel("output")
```

```python
plt.plot(x[:, 2], y, 'o')
plt.xlabel("input 3")
plt.ylabel("output")
```

```python
yhat = model.predict(x)
```

```python
print(y)
print("-------")
print(yhat)
```

```python
plt.plot(x[:, 0], y, 'o', label='True')
plt.plot(x[:, 0], yhat, 'o', label='Predicted')
plt.legend()
plt.xlabel("input 1")
plt.ylabel("output")
```

```python
plt.plot(x[:, 1], y, 'o', label='True')
plt.plot(x[:, 1], yhat, 'o', label='Predicted')
plt.legend()
plt.xlabel("input 2")
plt.ylabel("output")
```

```python
plt.plot(x[:, 2], y, 'o', label='True')
plt.plot(x[:, 2], yhat, 'o', label='Predicted')
plt.legend()
plt.xlabel("input 3")
plt.ylabel("output")
```

```python
plt.plot(y, yhat, 'o')
plt.xlabel('Real output')
plt.ylabel('Predicted output')
```
